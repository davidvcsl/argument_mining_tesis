```python
import pandas
import seaborn
```

```python
results = pandas.read_csv('../tmp/results.tsv', delimiter='\t')
results['attention'] = results.attention.replace('feature', 'context')
```

```python
results[:3]
```

```python
results.groupby(['activation']).agg(['max', 'min']).f1
```

```python
seaborn.factorplot(data=results[results.f1 > 0.65], x='activation', y='difference', kind='swarm', hue='attention')
```

```python
seaborn.distplot(results['difference'])
```

```python
seaborn.factorplot(data=results, x='activation', y='f1', kind='swarm', hue='attention')
```

```python
seaborn.factorplot(data=results[results.activation != 'softmax'], x='lstm', y='f1', kind='box', col='attention')
```

```python
seaborn.factorplot(data=results[results.activation != 'softmax'], x='batch', y='f1', kind='box', col='attention')
```

```python
seaborn.factorplot(data=results[results.activation != 'softmax'], x='dropout', y='f1', kind='box', col='attention')
```
